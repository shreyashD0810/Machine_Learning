```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
df.head()
```

```python
df.loc[5:10, 'sepal length (cm)'] = pd.NA
df[5:10]
```

```python
df['sepal length (cm)'][1].dtype
```

```python
df['sepal length (cm)'].fillna(df['sepal length (cm)'].mean(), inplace=True)
df[5:10]
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot(x='sepal length (cm)', data=df)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=df)
plt.title('Count of Each Species in the Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```
