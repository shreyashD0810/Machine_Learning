```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load and prepare data
iris_data = load_iris()
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df["species"] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)
```

```python

# Seaborn style
sns.set(style="whitegrid", palette="muted", font_scale=1.1)

# Value counts
print("Species Distribution:\n", df["species"].value_counts(), "\n")
```

```python

# Histogram
plt.figure(figsize=(10, 6))
df.hist(bins=20, figsize=(10, 6), color="#86bf91", edgecolor='black', grid=False)
plt.suptitle("Histogram of Iris Dataset", fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python

# Pairplot
sns.pairplot(df, hue="species", corner=True, diag_kind="kde",
             plot_kws={'alpha': 0.7, 's': 60, 'edgecolor': 'k'},
             palette="Set2")
plt.suptitle("Pairplot of Iris Features", fontsize=16, fontweight='bold', y=1.02)
plt.show()
```

```python

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.iloc[:, :-1], orient='h', palette="Set3")
plt.title("Boxplot of Iris Features", fontsize=14, fontweight='bold')
plt.xlabel("Value Range")
plt.tight_layout()
plt.show()
```

```python

# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.iloc[:, :-1].corr(), annot=True, cmap="YlGnBu", linewidths=0.5, linecolor='gray', fmt=".2f")
plt.title("Feature Correlation Heatmap", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
